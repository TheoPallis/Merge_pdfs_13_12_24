```python
import os
import glob

gramatia = r"C:\Users\pallist\Desktop\Desktop\ΤΡΕΧΟΝΤΑ\gramatia 2"
pdf_files = glob.glob(os.path.join(gramatia, "*.pdf"))

print(pdf_files)
```

```python
from PyPDF2 import PdfReader
import regex as re

for file in pdf_files:
    reader = PdfReader(file)
    text = ""

    for page in reader.pages:
        text += page.extract_text()

    unique_numbers = set(re.findall(r"(?<=Π)\d{7}", text))
    unique_numbers = sorted(unique_numbers)
    for code in unique_numbers:
        print(os.path.basename(file),code)
    # print(os.path.basename(file),unique_numbers)
```
